```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
"""
year = int from 20 to 23
quarter = int from 1 to 4
"""
def parse_page(year, quarter):
  r = requests.get(f"https://www.lyst.com/data/the-lyst-index/q{quarter}{year}/")
  page = BeautifulSoup(r.text, 'html.parser')

  # Brands
  brands_h3 = page.find_all(class_="hottest-brand-item__name")
  if len(brands_h3) == 0:
    brands_h3 = page.find_all(class_="hottest-brand-item__name h3")

  brands = [b.text.strip() for b in brands_h3]

  brand_table = pd.DataFrame({"place": range(1, 21), "brand": brands,
                              "year": [2000 + year]*20, "quarter": [quarter]*20})


  # Products
  products_h4 = page.find_all(class_="hottest-products__text")
  if len(products_h4) == 0:
    products_h4 = page.find_all(class_="hottest-products__text h4")

  products = [p.text.strip() for p in products_h4]

  if len(products) >= 20:
    products = products[:20]
    product_table = pd.DataFrame({"gender": ['women']*10 + ['men']*10, "place": list(range(1, 11))*2,
                                  "product": products, "year": [2000 + year]*20, "quarter": [quarter]*20})
  else:
    product_table = pd.DataFrame({"gender": ['all']*10, "place": range(1, 11),
                                  "product": products, "year": [2000 + year]*10, "quarter": [quarter]*10})


  return (brand_table, product_table)
```

```python
BRANDS = []
PRODUCTS = []

for year in range(19, 24):
  start = 2 if year == 19 else 1
  end = 4 if year == 23 else 5
  for quarter in range(start, end):
    brands, products = parse_page(year, quarter)
    BRANDS.append(brands)
    PRODUCTS.append(products)

brand_df = pd.concat(BRANDS)
product_df = pd.concat(PRODUCTS)
```
